#### Box plots
## pIC50 value

```python
plt.figure(figsize=(5.5, 5.5))

sns.boxplot(x = 'bioactivity_class', y = 'pIC50', data = df)

plt.xlabel('Bioactivity class', fontsize=14, fontweight='bold')
plt.ylabel('pIC50 value', fontsize=14, fontweight='bold')

plt.savefig('plot_ic50.pdf')
```

```python
box_values = df.groupby("bioactivity_class")["pIC50"].agg(
    Min="min",
    Q1=lambda x: x.quantile(0.25),
    Median="median",
    Q3=lambda x: x.quantile(0.75),
    Max="max"
)

print(box_values)
```

That corresponds roughly to nanomolar potency (very strong inhibitors).
The max 11.11 suggests an extremely potent outlier.

Almost all inactive compounds sit around pIC50 ≈ 4.
That corresponds to millimolar / very weak inhibition.

The distributions show clear separation with minimal overlap, suggesting reliable classification of bioactivity.

25% of your active molecules are extremely potent.

how classification is useful?


### Statistical analysis | Mann-Whitney U Test

```python
def mannwhitney(descriptor):

    from scipy.stats import mannwhitneyu

    # select needed columns
    selection = [descriptor, 'bioactivity_class']
    df_selected = df[selection]

    # split groups
    Active = df_selected[df_selected.bioactivity_class == 'Active'][descriptor]
    Inactive = df_selected[df_selected.bioactivity_class == 'Inactive'][descriptor]

    # run test
    stat, p = mannwhitneyu(Active, Inactive)

    # interpretation
    alpha = 0.05
    if p > alpha:
        interpretation = 'Same distribution (fail to reject H0)'
    else:
        interpretation = 'Different distribution (reject H0)'

    # results table
    results = pd.DataFrame({
        'Descriptor': descriptor,
        'Statistics': stat,
        'p-value': p,
        'alpha': alpha,
        'Interpretation': interpretation
    }, index=[0])

    return results
```

```python
mannwhitney('pIC50')
```

Hypothesis was there is no significant difference between active and inactive compounds
H0= 
H1= two groups are significantly different

Observation= theres difference
reject h0

The Mann-Whitney U test shows a significant difference in pIC₅₀ values between active and inactive compounds (p < 0.05), confirming that active molecules possess higher inhibitory potency.

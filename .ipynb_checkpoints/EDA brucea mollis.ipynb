```python
from rdkit import Chem
from rdkit.Chem import Descriptors
import pandas as pd

def lipinski(smiles):
    mol = Chem.MolFromSmiles(smiles)
    
    mw = Descriptors.MolWt(mol)
    logp = Descriptors.MolLogP(mol)
    hbd = Descriptors.NumHDonors(mol)
    hba = Descriptors.NumHAcceptors(mol)

    return pd.Series([mw, logp, hbd, hba])

df[["MW","LogP","HBD","HBA"]] = df["SMILES"].apply(lipinski)
df
```

```python
import numpy as np

df["standard_value"] = pd.to_numeric(df["standard_value"])

df["IC50_M"] = df["standard_value"] * 1e-9
df["pIC50"] = -np.log10(df["IC50_M"])
df
```

```python
def norm_value(input):
    norm = []
    
    for i in input:
        if float(i) > 100000000:
            i = 100000000
        norm.append(float(i))
        
    return norm

df["standard_value_norm"] = norm_value(df["standard_value"])
df
```

```python
df = df[df.bioactivity_class != "Intermediate"]
df
```

```python
def activity_level(pic50):
    
    if pic50 >= 6:
        return "High"
    
    elif pic50 <= 5:
        return "Low"
    
    else:
        return "Intermediate"

df["Activity_Level"] = df["pIC50"].apply(activity_level)
df
```

```python
median_value = df["pIC50"].median()

def activity_group(pic50):
    
    if pic50 >= median_value:
        return "Higher Activity"
    else:
        return "Lower Activity"

df["Activity_Group"] = df["pIC50"].apply(activity_group)
df
```

```python
def mannwhitney(descriptor):

    from scipy.stats import mannwhitneyu

    # select needed columns
    selection = [descriptor, 'Activity_Group']
    df_selected = df[selection]

    # split groups
    high = df_selected[df_selected.Activity_Group == 'High'][descriptor]
    low = df_selected[df_selected.Activity_Group == 'Low'][descriptor]

    # run test
    stat, p = mannwhitneyu(high, low)

    # interpretation
    alpha = 0.05
    if p > alpha:
        interpretation = 'Same distribution (fail to reject H0)'
    else:
        interpretation = 'Different distribution (reject H0)'

    # results table
    results = pd.DataFrame({
        'Descriptor': descriptor,
        'Statistics': stat,
        'p-value': p,
        'alpha': alpha,
        'Interpretation': interpretation
    }, index=[0])

    return results
```

```python
mannwhitney('pIC50')
```

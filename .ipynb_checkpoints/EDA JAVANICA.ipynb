```python
box_values = df.groupby("bioactivity_class")["pIC50"].agg(
    Min="min",
    Q1=lambda x: x.quantile(0.25),
    Median="median",
    Q3=lambda x: x.quantile(0.75),
    Max="max"
)

print(box_values)
```

```python
def mannwhitney(descriptor):

    from scipy.stats import mannwhitneyu

    # select needed columns
    selection = [descriptor, 'bioactivity_class']
    df_selected = df[selection]

    # split groups
    Active = df_selected[df_selected.bioactivity_class == 'Active'][descriptor]
    Inactive = df_selected[df_selected.bioactivity_class == 'Inactive'][descriptor]

    # run test
    stat, p = mannwhitneyu(Active, Inactive)

    # interpretation
    alpha = 0.05
    if p > alpha:
        interpretation = 'Same distribution (fail to reject H0)'
    else:
        interpretation = 'Different distribution (reject H0)'

    # results table
    results = pd.DataFrame({
        'Descriptor': descriptor,
        'Statistics': stat,
        'p-value': p,
        'alpha': alpha,
        'Interpretation': interpretation
    }, index=[0])

    return results
```

```python
mannwhitney('pIC50')
```

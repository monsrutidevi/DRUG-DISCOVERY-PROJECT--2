```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
target_query = target.search("coronavirus")
targets = pd.DataFrame.from_dict(target_query)
targets
```

here we are storing the new client in target
then searching for target
from the target we are making data frame using dict
printing targets to get it

```python
selected_target = targets.target_chembl_id[8] #in df targets target chemble id column was chosen and 8th index was taken
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type= "IC50")  # standard type ic50= how much of drug needed to inhibit 50% of the protein
```

in activity variable were storing activity of that target
in the activity were searching for the selected targets id whose standard type is IC50

```python
df= pd.DataFrame.from_dict(res)
df
```

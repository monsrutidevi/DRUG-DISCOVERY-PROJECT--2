```python

def lipinski(smiles_list, verbose=False):
    moldata = [Chem.MolFromSmiles(s) for s in smiles_list]  # convert all SMILES

    all_data = []

    for mol in moldata:
        desc_Molwt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)

        row = [desc_Molwt, desc_MolLogP, desc_NumHDonors, desc_NumHAcceptors]
        all_data.append(row)

        if verbose:
            print(f"Mol: {Chem.MolToSmiles(mol)} -> {row}")

    columnNames = ["MW", "LogP", "NumHDonors", "NumHAcceptors"]
    descriptors = pd.DataFrame(all_data, columns=columnNames)

    return descriptors

```
